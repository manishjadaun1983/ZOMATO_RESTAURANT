```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset1 = "../data/zomato - zomato.csv"
dataset2 = "../data/Country-Code - Sheet1.csv"
```

```python
df = pd.read_csv(dataset1)
df.head()
```

```python
# df.isnull().sum()

for col in df.columns:
    if df[col].isnull().any():
        print(col)
```

```python
dfCC = pd.read_csv(dataset2)
dfCC.head()
```

```python
Joineddf = df.merge(dfCC, left_on='Country Code', right_on='Country Code')
Joineddf.head()
```

```python
print(f'Number of unique countries in the dataset: {len(Joineddf["Country"].unique())}')
```

```python
print(f'Names of the countries in the dataset: ')
Joineddf['Country'].value_counts()
```

```python
country_names = Joineddf['Country'].value_counts().index
```

```python
country_vals = Joineddf['Country'].value_counts().values
```

```python
plt.pie(country_vals[:3], labels=country_names[:3], autopct='%1.1f%%')
plt.title('Top 3 countries with most restaurants')
plt.show()
```

```python
plt.bar(country_names[:3], country_vals[:3])
plt.title('Top 3 countries with most restaurants')
plt.show()
```
